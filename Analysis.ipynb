```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
#step 1 data import
data=pd.read_csv('D:\googleStore_DataAnalysis\googleplaystore.csv',encoding='latin1')
data.head(10)
```

```python
# prove random data in dataset
data.sample(10)
```

```python
data.info()
```

# STEP 2-> DATA CLEANING 

1-> CONVERT  DATA TYPES FOR BETTER ANALYSIS

```python
data['Price']=data['Price'].str.replace('Everyone','0')
data['Price']=data['Price'].str.replace(r'[^\d.]','',regex=True)
data['Price']=pd.to_numeric(data['Price'],errors='coerce')
```

```python
data['Installs']=data['Installs'].str.replace(r'[+,]','',regex=True)
data['Installs']=data['Installs'].str.replace('Free','0')
data['Installs']=pd.to_numeric(data['Installs'],errors="coerce")
```

```python
data['Last Updated']=pd.to_datetime(data['Last Updated'],errors='coerce')
```

```python
data.info()
```

2-> HANDALE MISSING VALUES

```python
data.isnull().sum()
```

```python
data['Rating'].fillna(data['Rating'].median(),inplace=True)
data['Type'].fillna(data['Type'].mode()[0],inplace=True)
data['Content Rating'].fillna(data['Content Rating'].mode()[0],inplace=True)
data['Last Updated'].fillna(data['Last Updated'].mode()[0],inplace=True)
data['Current Ver'].fillna(data['Current Ver'].mode()[0],inplace=True)
data['Android Ver'].fillna(data['Android Ver'].mode()[0],inplace=True)
```

3-> HANDLE DUPLICATE 

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(keep='first')
```

```python
data.drop_duplicates(inplace=True)
```

COPPING OF PRICE COLUMN OUTLIERS USING PERCENTILE METHOD

```python
upper_limit=data['Price'].quantile(0.95)
upper_limit
```

```python
data['Price']=data['Price'].clip(upper=upper_limit)
```

```python
# What is the average rating of apps on the play store ?
avarage_rating=data['Rating'].mean()
print(f'average rating of apps on the play store:  {avarage_rating:.2f}')
```

```python
#What percentage of app are free vs paid
total_apps=data.shape[0]
free_apps=data[data["Type"]=='Free'].shape[0]
paid_apps=data[data["Type"]=='Paid'].shape[0]


percentage_free_apps=(free_apps/total_apps)*100
percentage_paid_apps=(paid_apps/total_apps)*100
print(f'percentage of app are free:{percentage_free_apps : .2f}')
print(f'percentage of app are paid :{percentage_paid_apps : .2f}')

```
